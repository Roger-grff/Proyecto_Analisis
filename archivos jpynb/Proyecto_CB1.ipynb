```python
import pandas as pd

# Cargar el archivo CSV original
siniestros_df = pd.read_csv('2022_SINIESTROS_TRÁNSITO_BDD.csv', sep=';')

# Ver las primeras filas para revisión inicial
print("Vista preliminar del archivo original:")
print(siniestros_df.head())

# Eliminar filas duplicadas
siniestros_df.drop_duplicates(inplace=True)

# Eliminar filas completamente vacías
siniestros_df.dropna(how="all", inplace=True)

# Guardar el archivo limpio
siniestros_df.to_csv("siniestros_limpios.csv", index=False)

# Mostrar el resultado
print("\nArchivo limpio guardado:")
print(siniestros_df.head())
```

```python
import pandas as pd
from sqlalchemy import create_engine

# Leer el archivo limpio
df = pd.read_csv('sinestros_limpio.csv')

# Datos de conexión a MySQL
usuario = 'root'
contrasena = '1234'
host = 'localhost'
puerto = '3306'
base = 'trafico_siniestros_vehiculares'

# Crear conexión con SQLAlchemy
conexion = create_engine(f'mysql+pymysql://{usuario}:{contrasena}@{host}:{puerto}/{base}')

# Subir el DataFrame a MySQL (tabla: siniestros)
df.to_sql('siniestros', con=conexion, index=False)

print("Datos importados a MySQL.")
```

```python
#Analisis de Sentimiento de Títulos de Netflix
import pandas as pd

# Cargar el archivo CSV con títulos de Netflix
df = pd.read_csv('netflix_titles1.csv')

# Listas de palabras clave positivas y negativas
positive_words = ['fun', 'happy', 'love', 'great', 'excellent', 'adventure', 'inspiring']
negative_words = ['murder', 'crime', 'death', 'sad', 'abuse', 'tragedy', 'drug']

# Función simple de clasificación de sentimiento
def sentiment(text):
    text = str(text).lower()  # Convertir texto a minúsculas
    for word in positive_words:
        if word in text:
            return 'positive'
    for word in negative_words:
        if word in text:
            return 'negative'
    return 'neutral'

# Aplicar el análisis de sentimiento a la columna 'description'
df['sentiment'] = df['description'].apply(sentiment)

# Mostrar el resultado con título, descripción y sentimiento
print(df[['title', 'description', 'sentiment']])
```
